# Configuração da Biblioteca

```python
from openai import OpenAI
from google.colab import userdata

client = OpenAI(api_key=userdata.get('OPENAI_API_KEY'))
```

# Gerando Textos a partir de Imagem Online

Imagem para teste:<br/>
https://upload.wikimedia.org/wikipedia/commons/thumb/d/dd/Gfp-wisconsin-madison-the-nature-boardwalk.jpg/2560px-Gfp-wisconsin-madison-the-nature-boardwalk.jpg

```python
response = client.chat.completions.create(
    model="gpt-4o",
    messages=[
        {
            "role": "user",
            "content":[
                {"type":"text", "text": "Descreva o que há nesta imagem"},
                {"type":"image_url", "image_url": {
                    "url": "https://upload.wikimedia.org/wikipedia/commons/thumb/d/dd/Gfp-wisconsin-madison-the-nature-boardwalk.jpg/2560px-Gfp-wisconsin-madison-the-nature-boardwalk.jpg"
                }}
            ]
        }
    ],
    max_tokens=300
)
```

```python
print(response.choices[0].message.content)
```

# Gerando Textos a partir de Imagem Local

```python
import base64

def encode_image(image_path):
  with open(image_path, "rb") as image_file:
    return base64.b64encode(image_file.read()).decode("utf-8")
```

```python
base64_image = encode_image("/content/drive/MyDrive/_temp/image2.jpg")
```

```python
base64_image
```

```python
response = client.chat.completions.create(
    model="gpt-4o",
    messages=[
        {
            "role":"user",
            "content": [
                {"type":"text", "text": "Descreva o que há nesta imagem"},
                {"type": "image_url", "image_url": {
                    "url": f"data:image/jpeg;base64,{base64_image}"
                }}
            ]
        }
    ],
    max_tokens=300
)
```

```python
print(response.choices[0].message.content)
```
